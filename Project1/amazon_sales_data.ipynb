### Question 9.3 (5 marks)
Detect outliers in the discounted_price column using the IQR method.

**Outlier criteria:**
- Lower bound: Q1 - 1.5 * IQR
- Upper bound: Q3 + 1.5 * IQR

Count how many outliers exist and display 5 sample outliers.

**Hint:** Calculate bounds using quantiles and IQR, then filter

```python
# Your code here
q1 = df['discounted_price'].quantile(0.25)
q3 = df['discounted_price'].quantile(0.75)
iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

outliers = df[(df['discounted_price'] < lower) | (df['discounted_price'] > upper)]

print(f"Total outliers: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)")
print(f"Lower bound: ₹{lower:.2f}")
print(f"Upper bound: ₹{upper:.2f}")
print("\nSample outliers:")
outliers.nlargest(5, 'discounted_price')[['product_name', 'discounted_price']]
```

- Test with budget=1000, rating=4.0, category of your choice

**Hint:** Filter based on conditions, sort by multiple columns using `.sort_values()`

```python
# Your code here
def recommend_products(max_budget, min_rating, category=None):
    # Filter by budget and rating
    filtered = df[(df['discounted_price'] <= max_budget) & (df['rating'] >= min_rating)]
    
    # Filter by category if specified
    if category:
        filtered = filtered[filtered['category'] == category]
    
    # Sort by rating (desc) then discount (desc)
    recommendations = filtered.sort_values(['rating', 'discount_percentage'], 
                                          ascending=[False, False]).head(5)
    
    return recommendations[['product_name', 'rating', 'discounted_price', 'discount_percentage']]

# Test
print("Recommendations: Budget ₹1000, Rating >= 4.0")
recommend_products(max_budget=1000, min_rating=4.0)
```
