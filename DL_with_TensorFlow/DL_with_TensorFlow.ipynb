```python
df_new = df[["Account Length", "Area Code", "VMail Message", "Day Mins", 'Day Calls', 'Day Charge',
       'Eve Mins', 'Eve Calls', 'Eve Charge', 'Night Mins', 'Night Calls',
       'Night Charge', 'Intl Mins', 'Intl Calls', 'Intl Charge',
       'CustServ Calls' , "Churn?"]]
df_new
```

```python
y = df_new["Churn?"].apply(lambda x: 1 if x =="True." else 0)
y
```

# New section

```python
y.sum()
```

```python
x = df_new.drop("Churn?", axis = 1)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
x_train.head()
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python


x_train.shape
```

```python
# Creating the DL Model

# creating a DL model with sequential adding layers
DL_model1 = Sequential()
# creating hidden layer 1 and input layer
DL_model1.add( Dense(units = 32, activation = "relu", input_dim = len(x_train.columns) ))
# creating hidden layer 2
DL_model1.add(Dense(units = 16, activation = "relu"))
# creating hidden layer 3
#DL_model1.add(Dense(units = 32, activation = "relu"))
# creating output layer
DL_model1.add(Dense(units = 1, activation = "sigmoid"))
```

```python
# Build/Train the DL model

DL_model1.compile(loss="binary_crossentropy", optimizer="sgd", metrics=["accuracy"])


DL_model1.fit( x_train, y_train, epochs = 50, batch_size=32)
```

```python
y_train_hat = DL_model1.predict(x_train)
y_test_hat = DL_model1.predict(x_test)
```

```python
y_train_hat = [0 if val<0.7 else 1 for val in y_train_hat]
y_train_hat
y_test_hat = [0 if val<0.7 else 1 for val in y_test_hat]
y_test_hat
```

```python
accuracy_score( y_train, y_train_hat), accuracy_score( y_test, y_test_hat)
```
